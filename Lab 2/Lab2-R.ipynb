# Two Sample t-test - Divide dataset in two parts by company and perform t-test on adjusted earnings.


```r
vec = c('Marvel')
dx = data[data$Company %in% vec,]
dx
```

```r
vec = c('DC')
dy= data[data$Company %in% vec,]
dy
```

```r
boxplot(dx$Adjusted)
boxplot(dy$Adjusted)
```

```r
t.test(dx$Adjusted,dy$Adjusted)
```

```r
two_t = function(mx,my,nx,ny,sdx,sdy){
    sc = ((nx-1)*(sdx*sdx)+(ny-1)*(sdy*sdy))
    sc = sc/(nx+ny-2)
    sc = sqrt(sc)
    v = 1/nx+1/ny
    v = sqrt(v)
    deno = sc*v
    return (c((mx-my)/deno,sc))
}
```

```r
mx = mean(dx$Adjusted)
my = mean(dy$Adjusted)
nx = nrow(dx)-1
ny = nrow(dy)-1
sdx = sd(dx$Adjusted)
sdy = sd(dy$Adjusted)

t_val = two_t(mx,my,nx,ny,sdx,sdy)
cat("T = ",t_val[1])
sc = t_val[2]

td = qt(1-.05/2,nx+ny-2)
df = nx+ny-2

x=(mx-my)-((sc*td)*sqrt(1/nx+1/ny))
y=(mx-my)+((sc*td)*sqrt(1/nx+1/ny))
cat("Acceptance range = ",x,y)
```

# Paired t-test - Perform t-test over foreign and domestic columns.


```r
t.test(data$Foreign,data$Domestic, paired = TRUE)
```

```r
d =  data$Foreign - data$Domestic
```

```r
pair_test = function(m,n,sd){
    return (m/(sd/sqrt(n)))
}
```

```r
t_val = pair_test(mean(d),nrow(data),sd(d))
cat("T = ",t_val,"\n")

td = qt(1-.05/2,nrow(data))
x = mean(d)-td*(sd(d)/(sqrt(nrow(data))))
y = mean(d)+td*(sd(d)/(sqrt(nrow(data))))

cat("Confidence Interval = ",x,y)
```
